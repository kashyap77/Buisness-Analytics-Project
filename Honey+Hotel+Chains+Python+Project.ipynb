### 3.2.2 Create a new column "total_stay_duration" in the data frame not_cancel. (2 Marks)

Total_Stay_Duration is defined as the number of nights a customer stayed in the hotel including both weekends and weekdays

```python
not_cancel['Total_Stay_Duration'] = np.add(not_cancel['stays_in_weekend_nights'], not_cancel['stays_in_week_nights'])
## This code explains that we have created a new coloumn name Total stay Duration.so that we focus on customers who have longest stay in hotel.
```

```python
not_cancel.sort_values(['Total_Stay_Duration'],ascending=False)
# This code explains that we have taken data of maximum stay in descending order.
```

### 3.2.3 Create a new column "total_revenue" in the data frame not_cancel. (2 Marks)

Total_Revenue can be calculated by multiplying average rental revenue(adr) with total stay duration(column created in last question)

The average daily rate (adr) measures the average rental revenue earned for an occupied room per day. The operating performance of a hotel or other lodging business can be determined by using the ADR.

```python
not_cancel['Total_revenue'] = np.multiply(not_cancel['Total_Stay_Duration'], not_cancel['adr'])
# This code explains the total revenue generated by hotel rooms so that find out the customers through which highest revenue is generated.
```

```python
not_cancel.sort_values(['Total_revenue'],ascending=False)
# This code helps in sorting the data in descending order.
```
